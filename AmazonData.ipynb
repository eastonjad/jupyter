# Amazon Retail Data Analysis

## Read Retail Data Files

```python
import pandas as pd
```

```python
retailData1 = pd.read_csv("data/AmazonData/Retail.OrderHistory/Retail.OrderHistory.1.csv")
```

```python
retailData2 = pd.read_csv("data/AmazonData/Retail.OrderHistory/Retail.OrderHistory.2.csv")
```

```python
type(retailData1)
```

```python
retailData1
```

## Union Two Files into One

```python
dataFrames = [retailData1, retailData2]
```

```python
retailData = pd.concat(dataFrames)
```

```python
retailData
```

## Remove Commas and Change Column Types to Numeric

```python
retailData.describe()
```

```python
retailData.info()
```

```python
retailData.axes
```

```python
retailData["Price"] = pd.to_numeric(retailData["Price"].str.replace(',',''), errors='coerce')
```

```python
retailData["Item Subtotal"] = pd.to_numeric(retailData["Item Subtotal"].str.replace(',',''), errors='coerce')
```

```python
retailData.info()
```

```python
retailData.describe()
```

```python
retailData.sum()
```

## Convert Data Column

```python
retailData["Order Date"] = pd.to_datetime(retailData["Order Date"])
```

```python
retailData.info()
```

## Plot Graph

```python
%matplotlib inline
```

```python
retailData.groupby(pd.Grouper(key="Order Date", freq='Y'))["Item Subtotal"].sum().plot()
```

```python
retailData.groupby(pd.Grouper(key="Order Date", freq='Y'))["Item Subtotal"].sum().plot(kind="bar")
```
